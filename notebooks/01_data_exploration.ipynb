# 01 - Data Exploration

This notebook fetches stock data using `yfinance`, performs some basic cleaning, and displays interactive charts using Plotly. It also includes a GPU availability check (informational).

```python
# ensure torch is installed in this notebook environment
%pip install torch --quiet

# Check for local GPU (informational)
import torch # pyright: ignore[reportMissingImports]
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU name:", torch.cuda.get_device_name(0))

# Load libraries
import yfinance as yf
import pandas as pd
import plotly.express as px

# 1. Fetch data
ticker = 'AAPL'
data = yf.download(ticker, period='1y', auto_adjust=True)

# 2. FIX: Flatten the MultiIndex Columns
# This takes ('Close', 'AAPL') and turns it into just 'Close'
if isinstance(data.columns, pd.MultiIndex):
    data.columns = data.columns.get_level_values(0)

# 3. Reset index so 'Date' becomes a column instead of the index
data.reset_index(inplace=True)

# 4. Filter columns
df = data[['Date', 'Close', 'Volume']].copy()

# 5. Interactive closing price chart
# Now 'Date' and 'Close' are simple strings, so Plotly can find them
fig = px.line(df, x='Date', y='Close', title=f'{ticker} Closing Price (1 year)')
fig.show()

# 6. Interactive volume bar chart
fig2 = px.bar(df, x='Date', y='Volume', title=f'{ticker} Volume (1 year)')
fig2.show()
```
